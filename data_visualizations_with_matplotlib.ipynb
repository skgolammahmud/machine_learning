````python
# importing libraries
import matplotlib.pyplot as plt
import numpy as np

# show plots inside the notebook
%matplotlib inline
````

# A simple starter plot

````python
x_data = [1,2,3,4,5,6,7,8,9]
y_data = np.exp(x_data)
````

````python
plt.plot(x_data, y_data)
plt.title('Simple Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')
````

###### `x_data` and `y_data` can be lists, NumPy arrays, or a Pandas dataframes.

# Subplots
`plt.subplot` will display multiple plots on same figure.

````python
# plt.subplot(number of rows, number of columns, plot number)

plt.subplot(1,2,1)
plt.plot(x_data, y_data)
plt.title('Simple Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')

plt.subplot(1,2,2)
plt.plot(x_data, -y_data)
plt.title('Simple Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')

plt.show()
````

# Figure size

`plt.figure(figsize=(width, height))` can be used to adjust the figure size to the desired size.


````python
plt.figure(figsize=(10,8))
plt.plot(x_data, y_data)
plt.title('Simple Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')
````

##### Dot Per Inch (DPI) can be specified

````python
plt.figure(figsize=(8,5), dpi=100)
plt.plot(x_data, y_data)
plt.title('Simple Plot')
plt.xlabel('X Data')
plt.ylabel('Y Data')
````

# Saving figure
##### matplot figure can be saved in different formats such as PNG, JPG, PDF, etc.

````python
figure = plt.figure(figsize=(8,5))
figure.savefig('SavedFigure.png')
````

# Legend

````python
# Legend is added to figure by plt.legend()
plt.plot(x_data, y_data, label='y_data')
plt.plot(x_data, -y_data, label='-y_data')
plt.legend(loc='lower left')
````

#### Line Colors and Styles

color can be added to the line plots using color parameter, color codes(rgbcmyk), hex codes, and grayscale level.

```
Color codes (RGBCYMK):

'b'....blue
'g'....green
'r'....red
'c'....cyan
'm'....magenta
'y'....yellow
'k'....black
```
Hex codes:

Example: #00FF00' is green. Google any hex code color.

````python
plt.plot(x_data, y_data, color='red') # red
plt.plot(x_data, -y_data, color='y')  # yellow
plt.plot(x_data, 2*y_data, color='#00FF00')  # green
plt.plot(x_data, 3*y_data, color='0.8') # grayscale
````

#### Line styles and markers


# Bar Plot

Bar plots are used to display the relationship between two variables, one being mumeric and other being categorical.

````python
quantity = np.array([30,30,30,10])
fruits = ['Apple', 'Orange','Lemon', 'Pineapple']

plt.bar(fruits, quantity)
````

# Pie Charts

````python
plt.pie(quantity, labels=fruits);
````

# Histograms

Histograms are used to show the frequency distribution of the data values.

````python
fruits= ['Apple', 'Orange','Lemon', 'Pineapple','Orange', 'Pineapple']
plt.hist(fruits);
````

````python
data = np.random.randn(100)
plt.hist(data);
````
